# Практика 3

## Задание 1 Переводчик

Написать функцию-переводчик с русского на английский. Функция принимает строку текста на русском и возвращает строку на английском.

Использовать функцию в программе, которая при запуске входит в бесконечный цикл взаимодействия с пользователем. Программа запрашивает ввод от пользователя и если он не равен командам завершения работы ("выход", "exit"), то переводит введеный текст, иначе выходит из цикла.

Усложнение задания - написать транслятор на другие языки, такие как:
- Азбука Морзе
- [Leet](https://en.wikipedia.org/wiki/Leet)
- [Жаргон падонков, олбанский](https://ru.wikipedia.org/wiki/%D0%96%D0%B0%D1%80%D0%B3%D0%BE%D0%BD_%D0%BF%D0%B0%D0%B4%D0%BE%D0%BD%D0%BA%D0%BE%D0%B2)

## Задание 2 Чатбот с характером

Написать программу для интерактивного взаимодействия с чат-ботом.  Программа запрашивает ввод от пользователя и если он не равен командам завершения работы ("выход", "exit"), то передает запрос на обработку языковой модели.

Чатбот должен:
- Обладать ярко выраженным характером, проявляющимся в речевых оборотах, стилистике разговора, синтаксисе.
- Запоминать и использовать информацию из 3 последних сообщений пользователя.
