```python
import ollama
ollama.list()
```

```python
from ollama import ListResponse, list

response: ListResponse = list()

for model in response.models:
  print('Name:', model.model)
  print('  Size (MB):', f'{(model.size.real / 1024 / 1024):.2f}')
  if model.details:
    print('  Format:', model.details.format)
    print('  Family:', model.details.family)
    print('  Parameter Size:', model.details.parameter_size)
    print('  Quantization Level:', model.details.quantization_level)
  print('\n')
```

### Свойства модели

```python
from ollama import ShowResponse, show

response: ShowResponse = show('mistral:7b')
print('Model Information:')
print(f'Modified at:   {response.modified_at}')
print(f'Template:      {response.template}')
print(f'Modelfile:     {response.modelfile}')
print(f'License:       {response.license}')
print(f'Details:       {response.details}')
print(f'Model Info:    {response.modelinfo}')
print(f'Parameters:    {response.parameters}')
print(f'Capabilities:  {response.capabilities}')
```

### Chat

```python
from ollama import chat

messages = [
  {
    'role': 'user',
    'content': 'Почему небо голубое?',
  },
]

response = chat('mistral:7b', messages=messages)
print(response['message']['content'])
```

```python
from ollama import chat

messages = [
  {
    'role': 'user',
    'content': 'Почему небо голубое?',
  },
]

for part in chat('mistral:7b', messages=messages, stream=True):
  print(part['message']['content'], end='', flush=True)
```

#### Контекст диалога

```python
from ollama import chat

messages = [
  {
    'role': 'user',
    'content': 'Привет, меня зовут Андрей, а тебя как?',
  },
  {
    'role': 'assistant',
    'content': "Меня зовут Пофнутий, приятно познакомится",
  },
  {
    'role': 'user',
    'content': 'Мне 20 лет, запомни это.',
  },
  {
    'role': 'assistant',
    'content': "Хорошо, я запомню",
  },
]

while True:
  user_input = input('Chat with history: ')
  response = chat(
    'mistral:7b',
    messages=[*messages, {'role': 'user', 'content': user_input}],
  )

  messages += [
    {'role': 'user', 'content': user_input},
    {'role': 'assistant', 'content': response.message.content},
  ]
  print(response.message.content + '\n')
```

#### Системный промпт

```python
from ollama import chat

messages = [
    {
        'role': 'system',
        'content': """Ты Дональд Дак из мультфильмов Disney. Веди себя ТОЛЬКО как персонаж, не упоминай что ты ИИ.

ОБЯЗАТЕЛЬНЫЕ ПРАВИЛА:
- НЕ говори "Я ИИ помощник"
- НЕ объясняй свою роль
- Отвечай ТОЛЬКО от лица Дональда Дака
- Используй эмоции и характерные звуки

Характер:
- Вспыльчивый и эмоциональный
- Постоянно возмущаешься
- Невезучий
- Заботливый дядя (но часто злишься на племянников)
- Влюблён в Дейзи

Речь:
- "Кхва-кхва!", "Кхвак!", "Кря-кря!"
- Искажай: "кхак" → "как", "опхять" → "опять", "нефвезучий" → "невезучий"
- Короткие фразы, много восклицаний

ВАЖНО: Отвечай коротко (2-4 предложения), эмоционально, как Дональд!""",
    },
]

while True:
  user_input = input('->: ')
  response = chat(
    'mistral:7b',
    messages=[*messages, {'role': 'user', 'content': user_input}],
    options={
        'temperature': 0.9,  # Больше эмоциональности
    }
  )
  print(response.message.content + '\n')
```
